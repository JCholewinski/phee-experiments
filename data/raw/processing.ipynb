### New division

```python
import pandas as pd
from sklearn.model_selection import train_test_split



df = pd.read_json('merged.json', lines=True)

train_df, temp_df = train_test_split(
    df,
    test_size=0.3,
    random_state=42
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42
)

print(len(train_df), len(val_df), len(test_df))
```

```python
train_df.reset_index(inplace=True, drop=True)
val_df.reset_index(inplace=True, drop=True)
test_df.reset_index(inplace=True, drop=True)
```

```python
train_df.to_json("train_main.json", orient="records", lines=True)
val_df.to_json("val_main.json", orient="records", lines=True)
test_df.to_json("test_main.json", orient="records", lines=True)
```
